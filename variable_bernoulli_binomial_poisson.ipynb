## Tipos de variables aleatorias discretas

### Bernoulli

Sea $E$ un experimento y $\Omega$ su espacio muestral constituido únicamente por los eventos **Éxito** y **Fracaso**. Definamos una v.a. tal que $X=1$ cuando al llevarse a cabo el experimento el resultado es un **Éxito** y $X=0$ cuando ocurre lo contrario. Por lo tanto la función de masa de dicha variable se vería de la siguiente manera:

$$p(0) =  P\{X=0\} = 1 - p$$

$$p(1) =  P\{X=1\} = p$$

con $p,~0\leq p \leq 1$ la probabilidad de obtener un **Éxito**

Se dice entonces que una v.a. así descrita es de tipo Bernoulli con parámetro $(p)$(en honor del matemático suizo James Bernoulli (1654-1705).

### Binomial

Consideremos ahora el caso en el que llevamos a cabo $n$ pruebas independientes del experimento antes descrito y que el parámetro $p$ se preserva a lo largo de los intentos. Si ahora $X$ representa el conteo del número de **Éxitos** obtenidos en las $n$ repeticiones, decimos que $X$ es una v.a. binomial con parametros $(n,p)$. Por lo  tanto podemos concluir que una v.a. Bernoulli representa el caso particular de una v.a. Binomial con parámetros $(1,p)$.

**Ejercicio:**
1. ¿Cuál es la expresión de la esperanza de una v.a. Binomial y por ende, cuál es la de una Bernoulli?
2. ¿Cuál es la varianza de ambas v.a.?

La función de masa de una v.a. Binomial~$(n,p)$ está dada por:

$$p(i)={n\choose i}p^i(1-p)^{n-i}$$

Nótese que estamos considerando todas las combinaciones en las que $n$ intentos o ensayos o réplicas pueden ser éxitos y $n-i$ fracasos. Debido a que se trata de ensayos **independientes** podemos calcular la probabilidad de la ocurrencia conjunta de cualquier combinación de éxitos y fracasos mediante el producto de sus probabilidades individuales. 

Ejemplo:
¿Cuál es la probabilidad de obtener 2 éxitos en 3 ensayos?

$$p(2)={3\choose 2}p^2(1-p)^{2-3}=pp(1-p)+p(1-p)p+(1-p)pp$$

Recordemos por el triángulo de pascal, o por el desarrollo del binomio de Newton o porque lo recordamos de memoria de Álgebra Superior 1 que:

$${n\choose i}={n!\over{i!(n-i)!}}={{n(n-1)(n-2)...(2)(1)}\over{i!(n-i)(n-i-1)(n-i-2)...(2)(1)}}$$

Afirmación: $0\leq i\leq n$, por lo tanto $n-i$ debe de ser un factor en la expansión factorial de $n!$, así como los factores $(n-i-1)(n-i-2)...(2)(1)$. Entonces podemos escribir:

$${n \choose i} ={{n(n-1)...(n-i+1)}\over{i!}}$$

Esto será importante a continuación cuando conozcamos la v.a. Poisson.

Sin embargo, una vez conocida la función de masa de una v.a. Binomial, ¿cuál es su función de distribución acumulada?, si recordamos la definición ésta se vería de la siguiente manera:

$$P\{X\leq k\} = \sum_{i=1}^k{n\choose i}p^i(1-p)^{n-i}$$

### Poisson

Consideremos ahora una v.a. introducida por el matemático francés Simeón Denis Poisson (1781-1840), la cual presentó en un libro que buscaba aplicar la teoría de la probabilidad a la teoría de leyes y ensayos criminales. En realidad la distribución Poisson surge de pensar en un fenómeno binomial que tuviera una **$p$ relativamente pequeña y una $n$ muy grande**, de tal manera que el producto $\lambda = np$ o $p={\lambda \over n}$ sea moderado o manejable. Este supuesto está intimamente relacionado con la dificultad que representaba en ese entonces el calculo de factoriales grandes sobre todo si consideramos que aún no había computadoras. 

Del planteamiento anterior entonces:

$$p(i)={n\choose i}p^i(1-p)^{n-i}$$

$$p(i)={{n(n-1)...(n-i+1)}\over{i!}}p^i(1-p)^{n-i}$$

sustituyendo $p={\lambda \over n}$:

$$p(i)={{n(n-1)...(n-i+1)}\over{i!}}{\lambda^i \over n^i}(1-{\lambda \over n})^{n-i}$$

$$p(i)={{n(n-1)...(n-i+1)}\over{n^i}}{\lambda^i \over i!}(1-{\lambda \over n})^{n-i}$$

Recordemos que: $n\rightarrow \infty$, afirmación:

$${{n(n-1)...(n-i+1)}\over{n(n)...(n)}}=1$$

Más aún:

$$(1-{\lambda \over n})^{n-i}={{(1-{\lambda \over n})^{n}}\over {(1-{\lambda \over n})^{i}}}={e^{-\lambda}\over {(1-0)^{i}}}=e^{-\lambda}$$

Por lo tanto la función de masa de una v.a. Poisson sería:

$$p(i)={\lambda^i \over i!}e^{-\lambda}$$
